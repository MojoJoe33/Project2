```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
videogames = pd.read_csv("vgsales.csv")
```

```python
#creates dataframe without NaN and changes year data type
videogames_df = videogames.dropna().astype({'Year':'int'})
videogames_df
```

```python
videogames_plat_year = videogames_df[["Name", "Platform","Year","Genre","Publisher","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].groupby(['Platform','Year']).sum().reset_index()
videogames_plat_year
```

```python
#deletes rank column
videogames_df_all = videogames_df[["Name", "Platform","Year","Genre","Publisher","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].groupby(['Platform','Year','Publisher','Name','Genre']).sum().reset_index()
videogames_df_all
```

```python
year_sales = videogames_df[['Year', 'NA_Sales',"EU_Sales","JP_Sales","Other_Sales","Global_Sales"]].groupby(['Year']).sum().reset_index()
year_sales
```

```python
#saves as a csv files
videogames_df.to_csv('videogames_df.csv',index = False)
videogames_df_all.to_csv('videogames_sorted_df.csv',index = False)
videogames_plat_year.to_csv('videogames_platyear_df.csv',index = False)
year_sales.to_csv('year_sales.csv',index = False)

#saves as json
videogames_df.to_json('videogames_df.json',index = False, orient = 'table')
videogames_df_all.to_json('videogames_sorted_df.json',index = False, orient = 'table')
videogames_plat_year.to_json('videogames_platyear_df.json',index = False, orient = 'table')
year_sales.to_json('year_sales.json',index = False, orient = 'table')
```
